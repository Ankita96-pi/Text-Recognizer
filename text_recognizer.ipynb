#**Algorithm**

**Load the Image:**

Read the image from the file path specified in IMAGE_PATH.
Initialize an EasyOCR reader for English text detection.

**Text Detection:**
Use the EasyOCR reader to detect text in the image.
Store the results in the result variable.

**Bounding Box and Text Overlay:**

Extract the coordinates of the top-left and bottom-right corners of the bounding box for the first detected text.

Retrieve the detected text itself.

Draw a green rectangle around the detected text region on the image.

Overlay the detected text on the image using white color.

**Display the Modified Image:**

Show the modified image with the bounding box and overlaid text using plt.imshow(img) and plt.show().

**Iterate Through All Detected Text:**

For each detection in the result, repeat steps 3 and 4 to visualize all detected text regions.

```python
import easyocr
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
IMAGE_PATH = '0.jpg'
```

```python
reader = easyocr.Reader(['en'])
result = reader.readtext(IMAGE_PATH)
result
```

```python
top_left = tuple(result[0][0][0])
bottom_right = tuple(result[0][0][2])
text = result[0][1]
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
img = cv2.imread(IMAGE_PATH)
img = cv2.rectangle(img,top_left,bottom_right,(0,255,0),3)
img = cv2.putText(img,text,top_left, font, 0.5,(255,255,255),2,cv2.LINE_AA)
plt.imshow(img)
plt.show()
```

```python
plt.imshow(img)
plt.show()
```

```python
for detection in result:
    top_left = tuple(detection[0][0])
    bottom_right = tuple(detection[0][2])
    text = detection[1]
    print(text)
```
